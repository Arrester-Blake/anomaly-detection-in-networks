```python
def thresholder(col):
    factor =pd.cut(data[col],2)
    #interval
    interval =pd.value_counts(factor)
    threshold = interval.index[0].right
    print(col, 'threshold = ', threshold)
    dataset[col]=dataset[col].apply(lambda x: '{} < {}'.format(col,threshold)  if x< threshold else  '{} >= {}'.format(col, threshold))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
rf = RandomForestClassifier(n_estimators=500, criterion = 'gini', max_depth= 12)
```

```python
x_train, x_test, y_train, y_test = train_test_split(pd.get_dummies(dataset[[col for col in dataset.columns if col!='class']]), data['class'])
```

```python
rf.fit(x_train,y_train)
```

```python
pred = rf.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, pred))
```
